```python
import obspy
from obspy import UTCDateTime
import numpy as np
from obspy.core.inventory import Station, Channel, Site
```

```python
import matplotlib as plt
plt=__import__("matplotlib.pyplot")
plt.pyplot.subplots
```

```python
import os
my_path = os.path.abspath("./plots")
my_file = 'Figure_10_a'
```

```python
station_list = ["vasq", "jeng", "sati", "uclg","gllp"]
```

```python
fig, axs = plt.pyplot.subplots(5, 3, figsize=(30,15))
for i in range(0,len(station_list)):
    acc = np.loadtxt(station_list[i]+".bbp", comments="#")
    t = acc[:,0]
    ns = acc[:,1] #Column 2: North-south acceleration (cm/s/s) (+ is northward)
    ew = acc[:,2] #East-west acceleration (cm/s/s) (+ is eastward)
    ver = acc[:,3] #Up-down acceleration (cm/s/s) (+ is upward)
    axs[i,0].plot(t, ew)
    #axs[i,0].set_title("E-W", fontsize=20)
    axs[i,0].set_ylabel(station_list[i], fontsize=24)
    axs[i,1].plot(t, ns)
    #axs[i,1].set_title("N-S", fontsize=20)
    axs[i,2].plot(t, ver)
    #axs[i,2].set_title("ver", fontsize=20)
axs[0,0].set_title("E-W", fontsize=20)
axs[0,1].set_title("N-S", fontsize=20)
axs[0,2].set_title("ver", fontsize=20)
axs[4,0].set_xlabel("time[s]", fontsize=20)
axs[4,1].set_xlabel("time[s]", fontsize=20)
axs[4,2].set_xlabel("time[s]", fontsize=20)

fig.tight_layout()
fig.savefig(os.path.join(my_path, my_file))  
```
